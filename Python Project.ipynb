```python
df_numerized = df.copy()
```

```python
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
df.head()
```

```python
df_numerized.corr()
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(df_numerized.corr(method='pearson'),cmap='viridis',annot=True)
plt.title('Correlation Metrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
correlation_matrix = df_numerized.corr()
correlation_paris = correlation_matrix.unstack()
correlation_paris.head(10)
```

```python
sorted_pairs = correlation_paris.sort_values(ascending=True)
sorted_pairs.head(10)
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation.head(4)
```
